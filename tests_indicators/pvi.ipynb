```python
import yfinance as yf
```

```python
ticker = 'BTC-USD'

data = yf.download(ticker, auto_adjust=True, start = '2020-01-01')
```

```python
import sys
sys.path.append("../")
sys.path.append("../algo_trader")
```

```python
from algo_trader.lib.indicators.pvi import PVI

df = data.iloc[-300:]
pvi = PVI(255)
pvi.calculate(df)
pvi.plot()
```

```python
import matplotlib.pyplot as plt

# Armo los dos subgrficos con subplots
# figsize: Es la dimensin de toda la figura, que contiene los subgrficos
# nrows: Es la cantidad de subgrficos
# height_ratios: Se coloca la proporcin de alturas entre los subgrficos (la cantidad de elementos debe coincidir con las nrows ingresadas)
fig, ax = plt.subplots(figsize=(15,8), nrows=2, gridspec_kw={'height_ratios' : [2,1]})

df_pvi = pvi.calculate(df)
df_pvi['buy_signal'] = pvi.calc_buy_signals()
df_pvi['sell_signal'] = pvi.calc_sell_signals()
buySignals = df_pvi[df_pvi['buy_signal'] == True]
sellSignals = df_pvi[df_pvi['sell_signal'] == True]

ax[0].plot(df_pvi.index, df_pvi.Close)
ax[0].plot(buySignals.index, buySignals.Close, '^', label='Buy signal', color='green')
ax[0].plot(sellSignals.index, sellSignals.Close, 'v', label='Sell signal', color='red')
ax[0].legend(loc='upper left')
ax[0].grid()

ax[1].plot(df_pvi["PVI_EMA"], color='gray', linewidth=1)
ax[1].fill_between(df_pvi.index, df_pvi["PVI"], df_pvi["PVI_EMA"], where=df_pvi["PVI"] > df_pvi["PVI_EMA"], alpha=0.5, color='green')
ax[1].fill_between(df_pvi.index, df_pvi["PVI"], df_pvi["PVI_EMA"], where=df_pvi["PVI"] < df_pvi["PVI_EMA"], alpha=0.5, color='red')
ax[1].grid()
```
